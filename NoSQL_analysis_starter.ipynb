### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

# Get the latitude and longitude values from the Penang Flavours Restaurant
penang_flavours = establishments.find_one({'BusinessName': "Penang Flavours"})

degree_search = 0.01
latitude = penang_flavours['geocode']['latitude']
longitude = penang_flavours['geocode']['longitude']

# Establish limits on latitude and longitude
max_lat = latitude + degree_search
min_lat = latitude - degree_search
max_lng = longitude + degree_search
min_lng = longitude - degree_search

print(f"Latitude = {latitude} Longitude = {longitude}")
print(f"Lat (max, min): ({max_lat}, {min_lat})  Lng (max, min): ({max_lng}, {min_lng})")

# Set up query within limits to find restaurants with a rating of 5
query = { 'geocode.longitude' : {'$gte': min_lng, '$lte': max_lng},
          'geocode.latitude' : {'$gte': min_lat, '$lte': max_lat},
          'RatingValue' : 5 
        }

# Sort by hygiene score with lowest at the top
sort = [('scores.Hygiene', 1)]

# Run the query
results = establishments.find(query).sort(sort)

# Print the results
# Print number of results first
result_count = establishments.count_documents(query)
print(f"Restaraunts nearby: {result_count}")

# Pretty print first 5 results
for i in (range(5)):
    pprint(results[i])
```

```python
# Convert result to Pandas DataFrame
results_df = pd.DataFrame(results)
results_df.head(5)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene' : 0}}

# 2. Groups the matches by Local Authority
group_query = {'$group':{'_id': {'LocalAuthorityName': '$LocalAuthorityName'},
                         'count': {'$sum': 1}
                        }
              }

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {'count':-1}}
pipeline = [match_query, group_query, sort_values]
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of documents in result: ", len(results))

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print("Number of documents in result: ", len(results_df))

# Display the first 10 rows of the DataFrame
results_df.head(10)
```

```python
# Rename column headings, and reorder colums so the id column is first
results_df = results_df.rename(columns={"count": "number of establishments",
                                        "_id.LocalAuthorityName": "LocalAuthorityName"})
results_df = results_df[["LocalAuthorityName", "number of establishments"]]

# Display the first 10 rows of the updated DataFrame
results_df.head(10)
```
